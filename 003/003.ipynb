```python
tmp = []

operations = {
    '+' : lambda x, y: x + y,
    '-' : lambda x, y: x - y,
    '*' : lambda x, y: x * y,
    '//' : lambda x, y: x / y,
    '%' : lambda x, y: x % y,
    '**' : lambda x, y: x**y
}

def calc(op, val1, val2): 
    return int(operations[op](val1, val2))

with open("input_file.txt", "r") as input_file:
    for line in input_file:
        op, val1, val2 = line.split("    ")
        tmp.append(calc(op, int(val1), int(val2)))

with open("output_file.txt", "w") as output_file:
        output_file.writelines(str(line) + ',' for line in tmp[:-1])
        output_file.write(str(tmp[-1]))
```

```python
import_dict = {}
with open('import_file_2_1.txt', 'r') as import_file_1, \
open('import_file_2_2.txt', 'r') as import_file_2, \
open('output_file_2.txt', 'w') as output_file:
    for line in import_file_1:
        import_dict.update({line.strip() : import_file_2.readline().split()})
    output = ' '.join(key[int(import_dict[key][0]) : int(import_dict[key][1]) + 1] for key in import_dict)
    output_file.write(output)
```

```python
import json
import re

periodic_table = json.load(open("periodic_table.json", "r", encoding="utf-8"))

output_list = []
with open("import_file_3.txt", "r") as import_file_3:
    import_str = import_file_3.read().strip()
    import_list_3 = re.findall(r'[A-Z][^A-Z]*', import_str)
    for key in import_list_3:
        output_list.append(periodic_table[key])
print(''.join(output_list))
```
